## About Dataset
#### This dataset generated by respondents to a distributed survey via Amazon Mechanical Turk between 03.12.2016-05.12.2016. Thirty eligible Fitbit users consented to the submission of personal tracker data, including minute-level output for physical activity, heart rate, and sleep monitoring. Individual reports can be parsed by export session ID (column A) or timestamp (column B). Variation between output represents use of different types of Fitbit trackers and individual tracking behaviors / preferences.


VeryActiveMinutes in a smartwatch means the amount of time you spend engaging in physical activity that is considered vigorous or high intensity. This includes activities like running, cycling, or doing high-intensity workouts.


FairlyActiveMinutes in a smartwatch means the amount of time you spend engaging in physical activity that is considered moderate or low intensity. This includes activities like walking, gardening, or doing light workouts.


LightlyActiveMinutes in a smartwatch means the amount of time you spend engaging in light physical activity, such as casual walking, cleaning, or doing light household chores.


SedentaryMinutes in a smartwatch means the amount of time you spend being inactive, like sitting or lying down, while wearing the watch. This can help you stay active and avoid the negative health effects of prolonged sitting.
